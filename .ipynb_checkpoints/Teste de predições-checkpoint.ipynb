# Importando libs e dataset

```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import LabelEncoder
```

```python
new_records = pd.read_csv('new_records.csv')
immutable_data = pd.read_csv('new_records.csv')
```

```python
new_records
```

# Tratamento dos dados

```python
new_records['CategoricalAge'] = pd.cut(new_records['Age'], bins = 5)
```

```python
new_records['CategoricalFare'] = pd.qcut(new_records['Fare'], q = 4)
```

```python
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ''
```

```python
new_records['Title'] = new_records['Name'].apply(get_title)
```

```python
new_records['FamilySize'] = new_records['SibSp'] + new_records['Parch'] + 1
```

```python
new_records['IsAlone'] = 0
new_records.loc[new_records['FamilySize'] > 1, 'IsAlone'] = 1
```

```python
to_drop = ['PassengerId', 'Ticket', 'Name', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'FamilySize']
```

```python
new_records.drop(to_drop, axis = 1, inplace = True)
```

```python
le = LabelEncoder()
```

```python
for i in ['Sex', 'Embarked', 'CategoricalAge', 'CategoricalFare', 'Title']:
    new_records[i] = le.fit_transform(new_records[i])
```

# Carregando o modelo e fazendo predições

```python
import pickle
```

```python
model = pickle.load(open('titanic_model_svm.sav', 'rb'))
```

```python
model
```

```python
predictions = model.predict(new_records)
```

```python
immutable_data['Survived'] = predictions
```
